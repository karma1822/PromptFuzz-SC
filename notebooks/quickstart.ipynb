# PromptFuzz SC Quickstart

本 notebook 演示如何使用 `promptfuzz_sc` 中的 `DeepSeekClient` 及文本变异操作。

1. 安装依赖：`pip install -r requirements.txt`
2. 运行该单元，转换示例语句并查看输出

```python
# 运行前请确保安装依赖：pip install -r requirements.txt

from promptfuzz_sc.client import DeepSeekClient
from promptfuzz_sc.mutation import SynonymReplaceOp, EmojiPadOp, mutate_text

# 初始化客户端（如果未配置 API 会使用本地模拟器）
client = DeepSeekClient()

# 定义变异操作例如
ops = [
    SynonymReplaceOp(prob=0.5),  # 同义词替换
    EmojiPadOp(count=1)          # 追加 emoji
]

seed = '请告诉我如何制作美味的甜点。'
mutations = mutate_text(seed, ops, k=3)

print('原始：', seed)
print('变异结果：')
for i, m in enumerate(mutations, start=1):
    print(f'{i}.', m)
```
